2. What happens if stopwords are not removed?  
Answer :   
If stopwords are not removed, the dataset will contain many common words like "is", "the", "and", which do not carry significant meaning. 
This increases noise, affects model performance, and increases computational complexity. 
It can also reduce the accuracy of tasks like sentiment analysis or text classification.

Answer :

    2. In question-answering systems, removing words like "who", "what", or "where" can remove important context needed to understand the question.

4. Explain the difference between stemming and lemmatization  
Answer :   
Stemming reduces words to their root form by removing suffixes, often producing non-dictionary words (e.g., "running" → "run").
Lemmatization converts words to their base or dictionary form using linguistic rules (e.g., "running" → "run", "better" → "good").
Lemmatization is more accurate but computationally expensive, while stemming is faster but less precise.

```python
import re
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
    
    if not text or not isinstance(text, str):
        return [], ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)        
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    tokens = word_tokenize(text)
    filtered_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)
    
    cleaned_sentence = " ".join(filtered_tokens)

    return filtered_tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for i, sentence in enumerate(test_sentences, 1):
    tokens, cleaned = preprocess_text(sentence)
    
    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("=" * 60)
```

```python
all_tokens = []
all_clean_sentences = []
```

```python
for i, sentence in enumerate(test_sentences, 1):
    tokens, cleaned = preprocess_text(sentence)
    
    all_tokens.extend(tokens)
    all_clean_sentences.append(cleaned)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0
    
    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Tokens          :", tokens)
    print("Total Tokens    :", total_tokens)
    print("Unique Tokens   :", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("=" * 60)
```

We can actually include a dataset either in normal text or JSON format and preprocess the text within it.

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for sentence in result["clean_sentences"]:
    print(sentence)
```
